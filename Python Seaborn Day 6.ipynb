# Box Plot

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
F=sns.load_dataset('tips')
```

```python
F
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='day',y='total_bill',data=F,hue='time',showmeans=True , meanprops={"markeredgecolor":'w','marker':'+'}
            ,palette='vlag_r' )
plt.show()
```



# Cat Plot

```python
var=sns.load_dataset('tips').head(20)
```

```python
sns.catplot(x='tip', y='size', data=var,hue='sex',palette='vlag',kind='point')
plt.show()
```




# Seaborn Styling Plot

```python
Data=sns.load_dataset('tips')
```

```python
sns.set_style("darkgrid")
sns.barplot(x='day',y='total_bill',data=Data,saturation=1)
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(7,7))
sns.set_context('poster',font_scale=1
               
               )
sns.barplot(x='day',y='total_bill',data=Data,saturation=1)
sns.despine()
plt.show()
```
