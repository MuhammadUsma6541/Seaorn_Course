# Count Plot in Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset("tips")
```

```python
var
```

```python
sns.countplot(x="sex",data=var,hue="smoker",palette="gist_earth",saturation=0.5)
plt.show()
```

# Violin Plot

```python
sns.violinplot(y="total_bill",x='day',data=var,hue="time")
plt.show()
```

```python
sns.violinplot(y="total_bill",x='day',data=var,linewidth=2,palette='dark')
plt.show()
```

```python
sns.violinplot(y="total_bill",x='time',data=var,linewidth=2,palette='rocket',order=['Dinner',"Lunch"],saturation=0.5)
plt.show()
```

```python
sns.violinplot(y="day",x='total_bill',data=var,hue="sex",split=True,scale="width",inner='box')
plt.show()
```
