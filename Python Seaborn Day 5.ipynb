# Pair plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset("tips")
```

```python
var
```

```python
sns.pairplot(var,vars=['total_bill','tip'],hue="sex",markers=["*",'>'],hue_order=['Female','Male'])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=['Female','Male'],palette='dark',kind='kde',diag_kind='hist')#scatter,kde,hsit)
plt.show()
```

# Strip Plot

```python
data=sns.load_dataset("tips")
```

```python
data
```

```python
sns.stripplot(x="day",y="total_bill",data=data,palette='dark',hue='sex',linewidth=0.5,edgecolor='r',jitter=1)
plt.show()
```
