```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# Histogram

```python
y_1=sns.load_dataset("penguins").head()
```

```python
y_1
```

```python
sns.displot(y_1["flipper_length_mm"],bins=[180,185,190,195,200],kde=True,rug=True,color='g')
plt.show()
```

# BarPlot Method

```python
sns.barplot(x="bill_length_mm",y="body_mass_g",data=y_1,hue="sex")
plt.show()
```

# Scatter Plot Method

```python
y_1=sns.load_dataset("penguins")
y_1
```

```python
sns.scatterplot(x="bill_length_mm",y="body_mass_g",data=y_1,hue="sex",style="sex",size="sex",sizes=(40,50))
plt.show()
```

```python
m={"Male":"*","Female":'o'}
sns.scatterplot(x="bill_length_mm",y="body_mass_g",data=y_1,hue="sex",style="sex",size="sex",sizes=(70,100),palette="gist_ncar"
                 , alpha=1,markers=m)
plt.show()
```
